```python
%matplotlib notebook
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import scipy.stats as stats
import numpy as np
```

```python
# data files
AU_Region_Mobility_Report = "data/2020_AU_Region_Mobility_Report.csv"

# Read the mouse data and the study results
AU_Region_Mobility = pd.read_csv(AU_Region_Mobility_Report)
```

```python
AU_Region_Mobility.head()
```

There are many NAN values. we need to clean? or leave as empty for sorting later on?

```python
# Convert NaN values to empty String
nan_value =float("NaN")
AU_Region_Mobility.replace("",nan_value, inplace= True)
#only dropping NaN in sub_region_1 because if there is no state then there is no usable references
AU_Region_Mobility_cleaned = AU_Region_Mobility.dropna(subset=['sub_region_1'],inplace= True)
AU_Region_Mobility = pd.DataFrame(AU_Region_Mobility)
AU_Region_Mobility.head()
```

```python
# group by states use for all summaries
summary_statistics_df = AU_Region_Mobility.groupby('sub_region_1')

# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the movement types for each state
summary_retail_and_recreation = summary_statistics_df.agg(['mean','median','var','std','sem'])["retail_and_recreation_percent_change_from_baseline"]
#flatten the index
summary_retail_and_recreation.columns = [''.join(col).strip() for col in summary_retail_and_recreation.columns.values]

# convert to dataframe
summary_retail_and_recreation = pd.DataFrame(summary_retail_and_recreation)

# This gives us the summary on changes for retail_and_recreation_percent_change_from_baseline in 2020
#The standard deviation (sd) describes the variability between individuals in a sample; 
# The standard error of the mean (sem) describes the uncertainty of how the sample mean represents the population mean
# SEM is how far your estimate of the mean is likely to be from the true population mean.The lower the more accurate.
summary_retail_and_recreation
```
